```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score

import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
drive.mount('/content/drive')
```

```python
dataset = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/CampusX Tutorials/titanic.csv", usecols=["Age","Pclass", "Fare", "Survived"])
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
x_train, x_test, y_train, y_test = train_test_split(dataset.drop(columns=["Survived"]), dataset["Survived"], test_size=0.2, random_state=42)
```

```python
knn = KNNImputer(n_neighbors=2)
x_train_trf = knn.fit_transform(x_train)
x_test_trf = knn.transform(x_test)
```

```python
pd.DataFrame(x_train_trf, columns=x_train.columns)
```

```python
lr = LogisticRegression()
lr.fit(x_train_trf, y_train)
y_pred = lr.predict(x_test_trf)
accuracy_score(y_test, y_pred)
```
